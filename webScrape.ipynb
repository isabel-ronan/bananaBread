```python
def scrape_ingredients(pages, label):
    ingredients_lists = []
    labels = []
    for page in pages:
        try:
            response = requests.get(f"https://www.bbcgoodfood.com/recipes/{page}")
            soup = BeautifulSoup(response.text, "html.parser")
            text = soup.get_text(separator="\n", strip=True)
            text = text.split('Ingredients\nNutrition')[1].split('Nutrition:')[0]
            ingredients_lists.append(text)
            labels.append(label)
        except Exception as e:
            print(f"There was a problem loading recipe {page}. Exception: {e}")
        sleep(randint(2,10))
    return ingredients_lists, labels
```

```python
unit_to_grams = {
    "g": 1,
    "kg": 1000,
    "ml": 1,        # assume 1 ml = 1 g for water-like liquids
    "tsp": 5, # assume 1 tsp is 5 grams
    "tbsp": 15 # assume 1 tbsp is 15 grams
}

main_ingredients = [
    'flour', 
    'sugar',
    'egg',
    'butter',
    'oil',
    'baking powder',
    'baking soda',
    'bicarbonate of soda',
    'yeast',
    'starter',
    'milk',
    'water',
    'salt'
]
```

```python
import re
from fractions import Fraction
from typing import Dict, Any, List

def _parse_amount(s: str) -> float:
    s = s.strip()
    m = re.match(r'^(\d+)\s+(\d+)\/(\d+)$', s)
    if m:
        whole = int(m.group(1))
        num = int(m.group(2))
        den = int(m.group(3))
        return whole + float(Fraction(num, den))
    m = re.match(r'^(\d+)\/(\d+)$', s)
    if m:
        return float(Fraction(int(m.group(1)), int(m.group(2))))
    try:
        return float(s)
    except ValueError:
        raise ValueError(f"Unrecognised numeric format: {s!r}")

def get_ingredients(
    text: str,
    unit_to_grams: Dict[str, float] = {"g": 1, "kg": 1000, "ml": 1, "tsp": 5, "tbsp": 15},
    main_ingredients: List[str] = [ 'flour', 'sugar', 'egg', 'butter', 'oil', 'baking powder', 'baking soda', 'bicarbonate of soda', 'yeast', 'starter', 'milk', 'water', 'salt' ]
) -> Dict[str, Dict[str, Any]]:
    main_ingredients_lower = [mi.lower() for mi in main_ingredients]

    alias_map = {
        'g': 'g', 'gram': 'g', 'grams': 'g', 'g sachet': 'g',
        'kg': 'kg', 'kilogram': 'kg', 'kilograms': 'kg',
        'ml': 'ml', 'milliliter': 'ml', 'millilitre': 'ml', 'milliliters': 'ml', 'millilitres': 'ml',
        'tsp': 'tsp', 'teaspoon': 'tsp', 'teaspoons': 'tsp',
        'tbsp': 'tbsp', 'tablespoon': 'tbsp', 'tablespoons': 'tbsp',
    }

    alias_reverse = {k.lower(): v for k, v in alias_map.items()}

    lines = [l.strip() for l in text.splitlines() if l.strip()]

    ingredients: Dict[str, Dict[str, Any]] = {}
    i = 0
    qty_regex = re.compile(
        r'^\s*(\d+(?:\s+\d+/\d+)?|\d+/\d+|\d*\.\d+|\d+)\s*([a-zA-Z\.]+(?:\s+[a-zA-Z\.]+)*)?\s*$',
        re.I
    )

    while i < len(lines):
        line = lines[i]
        m = qty_regex.match(line)
        if not m:
            i += 1
            continue

        amount_str = m.group(1)
        unit_str = (m.group(2) or "").strip()

        next_line_consumed_for_unit = False
        if not unit_str and i + 1 < len(lines):
            nxt = lines[i + 1].split()
            if nxt:
                possible_unit = nxt[0]
                if re.fullmatch(r'[A-Za-z\.]+', possible_unit):
                    unit_str = possible_unit
                    next_line_consumed_for_unit = True

        try:
            amount_val = _parse_amount(amount_str)
        except ValueError:
            i += 1
            continue

        unit_norm = unit_str.lower().strip().rstrip('.')
        unit_key = alias_reverse.get(unit_norm) if unit_norm else None

        grams_val = None
        if unit_key and unit_key in unit_to_grams:
            grams_val = amount_val * unit_to_grams[unit_key]

        found_ingredient = None
        search_span = []
        if unit_str and unit_str in line:
            after_unit = line.split(unit_str, 1)[1].strip()
            if after_unit:
                search_span.append(after_unit)
        if next_line_consumed_for_unit:
            parts = lines[i + 1].split(None, 1)
            if len(parts) == 2:
                search_span.append(parts[1])
        if i + 1 < len(lines):
            search_span.append(lines[i + 1])
        if i + 2 < len(lines):
            search_span.append(lines[i + 2])

        for text_piece in search_span:
            for mi in main_ingredients_lower:
                if re.search(r'\b' + re.escape(mi) + r's?\b', text_piece, re.I):
                    found_ingredient = mi
                    break
            if found_ingredient:
                break

        if found_ingredient:
            prev = ingredients.get(found_ingredient)
            if prev is None:
                ingredients[found_ingredient] = {
                    'grams': grams_val,
                    'amount': amount_val,
                    'unit': unit_str or None
                }
            else:
                if prev['grams'] is not None and grams_val is not None:
                    prev['grams'] += grams_val
                else:
                    prev['grams'] = None
                prev['amount'] = (prev.get('amount', 0) or 0) + amount_val
        if next_line_consumed_for_unit:
            i += 2
        elif found_ingredient and (i + 1 < len(lines) and found_ingredient in lines[i + 1].lower()):
            i += 2
        else:
            i += 1

    return ingredients
```

```python
bread_results = []
for bread_ingredients in bread_list:
    bread_results.append(get_ingredients(bread_ingredients))

cake_results = []
for cake_ingredients in cake_list:
    cake_results.append(get_ingredients(cake_ingredients))

banana_results = []
for banana_ingredients in banana_list:
    banana_results.append(get_ingredients(banana_ingredients))
```

```python
def convert_to_nice_df(data):
    rows = []

    for recipe_dict in data:
        row = {}
        for ingredient, info in recipe_dict.items():
            # we store grams if available; fallback to amount if grams=None
            value = info["grams"] if info["grams"] is not None else info["amount"]
            row[ingredient] = value
        rows.append(row)

    df = pd.DataFrame(rows)
    return df
```

```python
bread_df = convert_to_nice_df(bread_results)
bread_df['label'] = 'bread'
cake_df = convert_to_nice_df(cake_results)
cake_df['label'] = 'cake'
banana_df = convert_to_nice_df(banana_results)
banana_df['label'] = 'banana'

overall_df = pd.concat([bread_df, cake_df, banana_df])
overall_df.to_csv('./data/allIngredients.csv')
```
